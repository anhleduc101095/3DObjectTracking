```
import torchvision
import os
import json

env_path = ""

root_dir = '{0}/tmp/'.format(env_path)
train_dataset = torchvision.datasets.CIFAR10(root=root_dir, train=True, download=True)
eval_dataset = torchvision.datasets.CIFAR10(root=root_dir, train=False, download=True)

idx = 0
batch_size = 1000

def build_image_text_dataset(root, folder, dataset):
    global idx
    labels = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

    if not os.path.exists(os.path.join(root, folder)):
        os.mkdir(os.path.join(root, folder))

    results = []
    batch_results = []
    batch_count = 1

    for img, label_idx in dataset:
        item = {}
        imgname = "IMG{:06d}.png".format(idx)
        filename = os.path.join(root, folder, imgname)
        idx += 1
        caption = 'this is a picture of {}.'.format(labels[label_idx])
        img.save(filename)
        item['caption_id'] = idx
        item['image_id'] = idx
        item['caption'] = caption
        item['image_path'] = filename
        results.append(item)

        if len(results) >= batch_size:
            batch_results.append(results)
            results = []
            batch_count += 1

    # Add remaining results if any
    if results:
        batch_results.append(results)

    return batch_results

def gen_caption_meta(root, name, meta_batches):
    for i, batch in enumerate(meta_batches, 1):
        save_path = os.path.join(root, f'{name}-{i}.json')
        with open(save_path, 'w') as fw:
            fw.write(json.dumps(batch, indent=4))
        print(f'Saved {save_path}')

train_batches = build_image_text_dataset(root_dir, 'train', train_dataset)
gen_caption_meta(root_dir, 'train', train_batches)

eval_batches = build_image_text_dataset(root_dir, 'eval', eval_dataset)
gen_caption_meta(root_dir, 'eval', eval_batches)
```

```
import towhee
from towhee import ops
from transformers import TrainingArguments
#step1
#get the operator, modality has no effect to the training model, it is only for the inference branch selection.
clip_op = ops.image_text_embedding.clip(model_name='clip_vit_base_patch16', modality='image').get_op()

env_path = ""
#step2
#trainer configuration, theses parameters are huggingface-style standard training configuration.
data_args = {
    'dataset_name': None,
    'dataset_config_name': None,
    'train_file': '{0}/tmp/train-1.json'.format(env_path),
    'validation_file': '{0}/tmp/eval-1.json'.format(env_path),
    'max_seq_length': 77,
    'data_dir': None,
    'image_mean': [0.48145466, 0.4578275, 0.40821073],
    "image_std": [0.26862954, 0.26130258, 0.27577711]
}

training_args = TrainingArguments(
    output_dir='./output/train_clip_exp',
    overwrite_output_dir=True,
    num_train_epochs=32,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    do_train=True,
    do_eval=True,
    eval_steps=1,
    remove_unused_columns=False,
    dataloader_drop_last=True,
    # Add any other arguments you need
)


model_args = {
    'freeze_vision_model': False,
    'freeze_text_model': False,
    'cache_dir': './cache'
}

#step3
#train your model
clip_op.train(data_args=data_args, training_args=training_args, model_args=model_args)
```
